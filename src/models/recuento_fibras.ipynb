```python
from torch_dataset import FiberGraphDataset
import torch
import numpy as np
ds = FiberGraphDataset(r'C:\Users\pablo\GitHub\tfm_prg\tractoinferno_graphs\testset')

# Crear un diccionario para contar las fibras de cada tipo
count_fibras = {}

# Recorrer el dataset y contar las fibras de cada tipo
for idx, suj in enumerate(ds):
    print(f"Procesando sujeto {idx+1}/{len(ds)}")
    labels = suj.y
    # Convertir a numpy
    labels = labels.to('cpu')
    labels = labels.numpy()
    # Contar las fibras de cada tipo de la forma mas eficiente
    unique, counts = np.unique(labels, return_counts=True)
    print(unique, counts)
    for tipo, count in zip(unique, counts):
        if tipo in count_fibras:
            count_fibras[tipo] += count
        else:
            count_fibras[tipo] = count

    

# Imprimir el resultado
for tipo, count in count_fibras.items():
    print(f"Tipo {tipo}: {count} fibras")
```
